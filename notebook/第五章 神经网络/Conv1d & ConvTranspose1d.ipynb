## Conv1d Layers



```python
import torch
x = torch.ones(1,1,7) 
conv_layer = torch.nn.Conv1d(in_channels=1, out_channels=1, 
                       kernel_size=3,stride=1,padding=0,bias=False)
conv_layer.weight.data = torch.ones(1,1,3) # set 1d conv kernel weights
y = conv_layer(x)

print(x, ' * ', conv_layer.weight.data, ' =', y)
```

## ConvTranspose1d Layers



X = [1x1, 1x1, 1x1] = [1, 1, 1]

```python
import torch
y = torch.ones(1,1,1) 
deconv_layer = torch.nn.ConvTranspose1d(in_channels=1, out_channels=1, 
                       kernel_size=3,stride=1,padding=0,bias=False)
deconv_layer.weight.data = torch.ones(1,1,3) # set 1d conv kernel weights
x = deconv_layer(y)

print(y, ' _deconv_ ', conv_layer.weight.data, ' =', x)
```



X = [1x1, 1x1 + 1x1, 1x1 + 1x1, 1x1] = [1, 2, 2, 1]
